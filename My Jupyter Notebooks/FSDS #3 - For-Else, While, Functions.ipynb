```python
# Q. Find if the given string is a palindrome or not?

##### SOLUTION: 1 #####

s = input("Enter a string: ")

if s==s[::-1]:
    print("The given string is Palindrome")
else:
    print("The given string is NOT Palindrome")
```

```python
# Q. Find if the given string is a palindrome or not using a FOR loop?

##### SOLUTION: 2 #####

s = input("Enter a string: ")

flag = 0

for i in range(len(s)):
    if s[i] == s[len(s)-1-i]:
        continue
    else:
        flag = 1
        break

if flag==0:
    print("The given string is Palindrome")
else:
    print("The given string is NOT Palindrome")
```

```python
# Q. Find if the given string is a palindrome or not using a FOR loop (without using flag)?

##### SOLUTION: 3 #####

s = input("Enter a string: ")

for i in range(len(s)):
    if s[i] != s[len(s)-1-i]:
        print("The given string is NOT Palindrome")
        break
else:
    print("The given string is Palindrome")
```

```python
# Q. Find the maximum value from a nested dictionary for each keys in the given dictionary

##### SOLUTION: 1 #####

d1 = {
    "apdaga": {
        "a":12,
        "b":65,
        "c":7
    },
    "dumpbox":{
        "d":34,
        "e":83,
        "f":45
    }
}

# print(d1)

for k,v in d1.items():
    m = max(v.values())
    print(k, m)        
```

```python
# Q. Find the maximum value from a nested dictionary for each keys in the given dictionary
# without using built-in max function

##### SOLUTION: 2 #####

d1 = {
    "apdaga": {
        "a":12,
        "b":65,
        "c":7
    },
    "dumpbox":{
        "d":34,
        "e":83,
        "f":45
    }
}

# print(d1)

for k,v in d1.items():
    m = 0
    for i in v.values():
        if i>m:
            m=i
            
    print(k, m)
```

```python
# Q. Find the maximum integer out of all the values from a given dictionary d1

##### SOLUTION: 1 #####

d1 = {
    "apdaga": {
        "a":12,
        "b":65,
        "c":7
    },
    "dumpbox":{
        "d":34,
        "e":83,
        "f":45
    },
    "g":34,
    "h":[35,6,2,9,3],
    "i":(65,34,2),
    "k":"akshay"
}

# print(d1)

out=[]

for k,v in d1.items():
    if type(v) == dict:
        out.extend(list(v.values()))
    elif type(v) == int:
        out.append(v)
    elif type(v) == float:
        out.append(v)
    elif type(v) == list:
        out.append(max(v))
    elif type(v) == tuple:
        out.append(max(v))
    else:
        pass

# print(out)        
print(max(out))        
```

# Q. Write a function Print the pyramid structure as shown below in python for given number of rows/columns

##### * 
##### * * 
##### * * * 
##### * * * * 
##### * * * * * 

```python
##### SOLUTION: 1 #####
def prinPyramid(r):
    for i in range(r):
        for j in range(i+1):
            print("*", end="")
        print("")
```

```python
prinPyramid(5)
```

```python
##### SOLUTION: 2 #####
def prinPyramid2(r):
    for i in range(r):
        print("*"*(i+1))
```

```python
prinPyramid2(7)
```

# How to write fuctionality of the function (as we can see for built-in functions)

```python
def prinPyramid2(r):
    '''
    This function will print the right angle traingle
    made up of "*" (astrix) for the given input row/columns (r).
    '''
    for i in range(r):
        print("*"*(i+1))
```
